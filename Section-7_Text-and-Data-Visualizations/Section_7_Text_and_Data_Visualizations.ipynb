```python
# 1) Exercise - Write out alphabets and write which ordinal position that alphabet is in.
import string

alphabet_string = string.ascii_lowercase

alphabet_list = list(alphabet_string)

for i in range(len(alphabet_list)):
  if i == 0 or i == 20:
    ordind = 'st'
    #print('%s is the %gst letter of the alphabet' %(alphabet_list[i],i+1) )
  elif i == 1 or i == 21:
    ordind = 'nd'
    #print('%s is the %gnd letter of the alphabet' %(alphabet_list[i],i+1) )
  elif i == 2 or i == 22:
    ordind = 'rd'
    #print('%s is the %grd letter of the alphabet' %(alphabet_list[i],i+1) )
  else:
    ordind ='th'
  
  print('%s is the %g%s letter of the alphabet' %(alphabet_list[i],i+1,ordind))
```
