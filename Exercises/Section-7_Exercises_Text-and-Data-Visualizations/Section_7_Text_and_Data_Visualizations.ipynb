```python
# 4) Exercise - Graph a bunch of x^n lines. Make a colour gradient using them. (Much easier to visualize what I'm talking 
# talking about by looking at the results)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

fig,ax = plt.subplots()

for i in np.linspace(0,1,50):
  plt.plot(x,x**i, color=[i/5,i,i/2])
  

#ax.set_aspect(1/ax.get_data_ratio())

plt.show()
```

```python
# 5) Exercise -  Reproduce a graph # hint, arrow = wedge

x = np.linspace(-4,4,100)

y = x**2 + np.cos(x)*10

minimum = np.argmin(y)
txt = 'min: (%0.1f,%0.2f)'%(x[minimum],y[minimum])
print(txt) 
print(minimum)
fig,ax = plt.subplots()

ax.plot(x,y,'mo-',markerfacecolor = 'purple')
ax.annotate(txt,(x[minimum],y[minimum]), arrowprops=dict(arrowstyle ='wedge', color ='purple'), xytext =(-3,8)  )

ax.set_xlabel('x')

ax.set_ylabel('f(x)=x^2 + 10*cos(x)')

plt.show()
```

```python
# 6) Exercise - Create a regression plot of data df using seaborn

import numpy as np
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt

n = 200
D = np.zeros((n,2))
D[:,0] = np.linspace(0,10,n) +np.random.randn(n)
D[:,1] = D[:,0]**2 +np.random.randn(n)*10

df = pd.DataFrame(data = D, columns = ['var1','var2'])

sns.regplot(df.columns[0], df.columns[1], data = df)

plt.title(f'Regression of {df.columns[0]} on {df.columns[1]}')

plt.show()
```

```python
# 7) Exercise
# 1) Create a Hilbert Matrix (10x10) H[i,j] = 1 / (i+j-1)
import numpy as np


q=10
a=10
matrix = np.zeros ( [q, a] )


matrix

for i in range(1,q+1):
  for j in range(1,a+1):
    matrix[i-1,j-1] = 1/(i + j-1)


matrix

# 2) Visualize the Hilbert Matrix using seaborn 
import seaborn as sns

print(sns.heatmap(matrix))
print(np.round(matrix,2))
```
