```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from google.colab import files

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

df = pd.read_csv(url, sep =';')

df
```

```python
# removing the rows with outliers in total sulfur dioxide 

df = df[df ['total sulfur dioxide']<200]

```

```python
sns.pairplot(df[['alcohol','pH','quality']],kind='reg',hue='quality')
plt.show()
```

```python
# t-test of volatile acidity on quality 3 vs. 8

#grabbing the data 

x= df['volatile acidity'][df['quality'] == 3]
y = df['volatile acidity'][df['quality'] == 8]

# doing the t-test

ttest= stats.ttest_ind(x,y)

#plotting the data 

plt.plot(np.random.randn(len(x))/30, x, 'o',
         np.random.randn(len(y))/30+1, y, 'o')
plt.xlim([-1,2])
plt.xticks([0,1], labels=['Quality 3','Quality 8'])
plt.ylabel('volatile acidity')
plt.title(f't={ttest[0]:.2f}, p={ttest[1]:.6f}')

plt.show()
```

```python


qualcounts = np.zeros(6)

n=0
for i in range (3,9):
 qualcounts[n] = len(df[df['quality']==i])
 n += 1

qualcounts
plt.bar(range(3,9),qualcounts)
plt.xlabel('Quality Rating')
plt.ylabel('Counts')

plt.show()
```

```python
# repeating the t-test but using quality 3/4 vs. 7/8

#grabbing the data 

x = df['volatile acidity'][(df['quality'] == 3) | (df['quality'] == 4)]
y = df['volatile acidity'][(df['quality'] == 8) | (df['quality'] == 7)]

# doing the t-test

ttest= stats.ttest_ind(x,y)

#plotting the data 

plt.plot(np.random.randn(len(x))/30, x, 'o',
         np.random.randn(len(y))/30+1, y, 'o',
         markeredgecolor = 'k')
plt.xlim([-1,2])
plt.xticks([0,1], labels=['Quality 3/4','Quality 7/8'])
plt.ylabel('volatile acidity')
plt.title(f't={ttest[0]:.2f}, p={ttest[1]:.10f}')

plt.show()
```

```python
# Multiple Regression

import statsmodels.api as sm

dep_var  = df['quality']
ind_vars = df.loc[ : , df.columns != 'quality']

ind_vars = sm.add_constant(ind_vars)

model = sm.OLS(dep_var,ind_vars).fit()

print(model.summary())
```

```python
#finding all the variables with significant p-values 

sig_cols = list(model.pvalues[model.pvalues<0.05].keys())
[print (i) for i in sig_cols];

sig_cols.append('quality')

sns.pairplot(df[sig_cols], kind='reg',hue = 'quality' )
```

```python
#Logistic Regression

binthresh = np.mean(df['quality'])

df['binquality'] = df['quality'] > binthresh

df
```

```python
Xcols = []

for key in df.keys():
  if key not in ['quality','binquality']:
    Xcols.append(key)

model = sm.Logit(df['binquality'],df[Xcols])

results = model.fit(method ='newton')
results.summary()

```

```python
sig_colsL = list(results.pvalues[results.pvalues<0.05].keys())

print('Significant Predictors from the Logistic Regression')


[print("   "+ i) for i in sig_colsL];

print(" ")

print('Significant Predictors from the Standard Regression')


[print("   "+ i) for i in sig_cols[:-1]];
```
